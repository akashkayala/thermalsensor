```python
def draw_boxes(arr, boxes):
    for box in boxes:
            p1 = tuple(box[0:2].astype(np.int))
            p2 = tuple(box[2:4].astype(np.int))
            arr_bb = cv2.rectangle(arr, p1, p2, color=(255, 255, 255), thickness=2)
            
    return arr
```

```python
from PIL import Image
import cv2
from facenet_pytorch import MTCNN, InceptionResnetV1

mtcnn = MTCNN(
    image_size=160, margin=0, min_face_size=20,
    thresholds=[0.6, 0.7, 0.7], factor=0.709, post_process=True,
    device='cuda')

img = Image.open("../captured_frames/test_images/1594774387vis.jpg")

```

```python
arr = cv2.resize(np.array(img), (80,60))
%time boxes, _ = mtcnn.detect(arr)
arr_bb = draw_boxes(arr, boxes)
plt.imshow(arr_bb)
```

```python
arr = cv2.resize(np.array(img), (800,600))
%time boxes, _ = mtcnn.detect(arr)
arr_bb = draw_boxes(arr, boxes)
plt.imshow(arr_bb)
```
